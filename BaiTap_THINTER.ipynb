```python
#Bài 2: Thiết kế giao diện gồm 2 Entry, các Button tương ứng với 4 thao tác: +, -, *, /. 
#Viết code để khi bấm vào button thì kết quả được hiển thị ở Label tương tứng với button được bấm. 
# (Bấm + thì hiển thị kết quả cộng 2 số được nhập vào từ 2 Entry, tương tự với các thao tác khác)

from tkinter import *

root = Tk()
root.title("Phép tính")
root.geometry('400x300')
root.eval('tk::PlaceWindow . center')


def plus(event):
    try:
        number_start = int(number1.get())
        number_end = int(number2.get())
        result = number_start + number_end
        show["text"] = result
        show["fg"] = "green"
    except:
        show["text"] = "ERROR"
        show["fg"] = "red"
        

def minus(event):
    try:
        number_start = int(number1.get())
        number_end = int(number2.get())
        result = number_start - number_end
        show["text"] = result
        show["fg"] = "green"
    except:
        show["text"] = "ERROR"
        show["fg"] = "red"
        
        
def multiply(event):
    try:
        number_start = int(number1.get())
        number_end = int(number2.get())
        result = number_start * number_end
        show["text"] = result
        show["fg"] = "green"
    except:
        show["text"] = "ERROR"
        show["fg"] = "red"
        
        
def division(event):
    try:
        number_start = int(number1.get())
        number_end = int(number2.get())
        if number_end == 0:
            raise ValueError("Không thể thực hiện phép toán")
        else:
            result = number_start / number_end
            show["text"] = result
            show["fg"] = "green"
    except:
        show["text"] = "ERROR"
        show["fg"] = "red"
        
        
so1 = Label(root, text = "Số thứ nhất:").place(x = 40, y = 40)
so2 = Label(root, text = "Số thứ hai:").place(x = 40, y = 80)
ket_qua = Label(root, text = "Kết quả phép tính:").place(x = 40, y = 120)

number1 = Entry(root, text = "")
number1.place(x = 110, y = 40)
number2 = Entry(root, text = "")
number2.place(x = 110, y = 80)

show = Label(root, text = "")
show.place(x = 150, y =120)

#
button_plus = Button(root,text = '+')
button_plus.place(x = 30, y = 170, width=40, height=30)
button_plus.bind('<Button-1>', plus)
number1.focus()

button_minus = Button(root, text = '-')
button_minus.place(x = 90, y = 170, width=40, height=30)
button_minus.bind('<Button-1>', minus)

button_multiply = Button(root, text = 'x')
button_multiply.place(x = 150, y = 170, width=40, height=30)
button_multiply.bind('<Button-1>', multiply)

button_division = Button(root, text = ':')
button_division.place(x = 210, y = 170, width=40, height=30)
button_division.bind('<Button-1>', division)


root.mainloop()


```
